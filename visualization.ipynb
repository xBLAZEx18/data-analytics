```python
data=[1,2,3,4,6,8,10,12]
fig, ax = plt.subplots(figsize=(5,3))
ax.set_xlabel('no of elements',fontdict ={'fontsize':10,'color':'red'})
ax.set_ylabel('Y - value')
plt.plot(data)
```

```python
data=[1,2,3,4,6,8,10,12]
fig, ax = plt.subplots(figsize=(5,3))
ax.set_xlabel('no of elements',fontdict ={'fontsize':10,'color':'red'})
ax.set_ylabel('Y - value')
ax.grid(True)
ax.set_title("basics of visualization",loc='right')
plt.plot(data)
```

```python
data=[1,2,3,4,6,8,10,12]
fig, ax = plt.subplots(figsize=(5,3))
ax.set_xlabel('no of elements',fontdict ={'fontsize':10,'color':'red'})
ax.set_ylabel('Y - value')
ax.grid(True)
ax.set_title("basics of visualization",loc='right')
plt.plot(data, marker='X',markersize=10,markeredgecolor='black',markerfacecolor='purple')
```

```python
data=[1,2,3,4,6,8,10,12]
data2=[22,23,34,55,77,33,12]
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(data,marker='X',markersize=10,markeredgecolor='black',markerfacecolor='yellow',linestyle=':')
ax.plot(data2,marker='X',markersize=10,markeredgecolor='black',markerfacecolor='red',linestyle='--',color='green',linewidth='2',alpha=0.5)
```

```python

data=[1,2,3,4,6,8,10,12]
data2=[22,23,34,55,77,33,12]
fig, (ax1,ax2) = plt.subplots(figsize=(10,3),ncols=2)
ax1.plot(data,marker='X',markersize=10,markeredgecolor='black',markerfacecolor='yellow',linestyle=':')
ax2.plot(data2,marker='X',markersize=10,markeredgecolor='black',markerfacecolor='red',linewidth='2',alpha=0.5)
```
